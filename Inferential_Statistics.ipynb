# **Problem 2 :- Breaking Strength of gunny bags**

````python
# import the required function
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt

# From the problem statement we have mean and standard deviation.
mu = 5
sigma = 1.5
print("The mean is", mu, "kg per sq. centimeter.")
print("The standard deviation is", sigma, "kg per sq. centimeter.")
````

````python
# X value plotting
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
# Normal distribution pdf
y = norm.pdf(x, mu, sigma)
# Plotting
plt.plot(x,y)
plt.title('Normal Distribution')
plt.xlabel('x')
plt.ylabel('pdf')
plt.legend()
plt.grid(True)
plt.show
````

**2.1 Calculating proportion of the gunny bags have a breaking strength of less than 3.17 kg per sq cm.**

````python
#find the cumulative probability, norm.cdf() calculates the cumulative probability
x1=3.17
less_than_x1 = norm.cdf(3.17, mu, sigma)
print("The proportion of the gunny bags have a breaking strength of less than 3.17 kg per sq cm is", round(less_than_x1, 4))
````

````python
#plotting the distribution
plt.plot (x, y)
plt.axvline(x1, color='red')
x2=np.linspace(x.min(),x1)
plt.fill_between(x2, norm.pdf(x2, mu, sigma), color='red')
plt.title('Normal Distribution')
plt.xlabel('x')
plt.ylabel('pdf')
plt.legend()
plt.grid(True)
plt.show
````

**2.2 Calculating proportion of the gunny bags have a breaking strength of at least 3.6 kg per sq cm.**


````python
#find the cumulative probability, norm.cdf() calculates the cumulative probability
x3=3.6
at_least_x3 =1- norm.cdf(x3, mu, sigma)
print("The proportion of the gunny bags have a breaking strength of at least 3.6 kg per sq cm is", round(at_least_x3, 4))
````

````python
#plotting the distribution
plt.plot (x, y)
plt.axvline(x3, color='red')
x4=np.linspace(x3,x.max())
plt.fill_between(x4,norm.pdf(x4, mu, sigma), color='red')
plt.title('Normal Distribution')
plt.xlabel('x')
plt.ylabel('pdf')
plt.legend()
plt.grid(True)
plt.show
````

**2.3 Calculating proportion of the gunny bags have a breaking strength between 5 and 5.5 kg per sq cm.**

````python
#find the cumulative probability, norm.cdf() calculates the cumulative probability
x5=5
x6=5.5
between_x5_and_x6 = norm.cdf(x6, mu, sigma) - norm.cdf(x5, mu, sigma)
print("The proportion of the gunny bags have a breaking strength between 5 and 5.5 kg per sq cm is", round(between_x5_and_x6, 4))
````

````python
#plotting the distribution
plt.plot (x, y)
plt.axvline(x5, color='red')
plt.axvline(x6, color='red')
x7=np.linspace(x5,x6)
plt.fill_between(x7,norm.pdf(x7, mu, sigma), color='red')
plt.title('Normal Distribution')
plt.xlabel('x')
plt.ylabel('pdf')
plt.legend()
plt.grid(True)
plt.show
````

**2.4 Calculating proportion of the gunny bags have a breaking strength NOT between 3 and 7.5 kg per sq cm.**

````python
#find the cumulative probability, norm.cdf() calculates the cumulative probability
x8=3
x9=7.5
not_between_x8_and_x9 = 1-(norm.cdf(x9, mu, sigma) - norm.cdf(x8, mu, sigma))
print("The proportion of the gunny bags have a breaking strength  not between 3 and 7.5 kg per sq cm is", round(not_between_x8_and_x9, 4))
````

````python
#plotting the distribution
plt.plot (x, y)
plt.axvline(x8, color='red')
plt.axvline(x9, color='red')
x10=np.linspace(x.min(),x8)
plt.fill_between(x10,norm.pdf(x10, mu, sigma), color='red')
x11=np.linspace(x9,x.max())
plt.fill_between(x11,norm.pdf(x11, mu, sigma), color='red')
plt.title('Normal Distribution')
plt.xlabel('x')
plt.ylabel('pdf')
plt.legend()
plt.grid(True)
plt.show
````

**One sample t-test**

**3.1 Zingaro has reason to believe that the unpolished stones may not be suitable for printing. Do you think Zingaro is justified in thinking so?**

###Step 1: Define null and alternative hypothesis

$H_0:$  The mean Brinell's hardness of unpolished stones is greater than or equal to 150.

$H_a:$ The mean Brinell's hardness of unpolished stones is less than 150.

Let $\mu$ be the mean Brinell's hardness of stones.

Mathematically, the above formulated hypotheses can be written as:

>$H_0: \mu \ge 150 $ <br>
$H_a: \mu < 150 $

### Step 2: Select Appropriate test

This is a case of a one-tailed test for the significance of a single mean. As the population standard deviation is unknown, a one-sample t-test will be appropriate to test the formulated hypotheses.

In a **one sample t-test**, we compare a sample mean with a hypothesized population mean to check if the sample mean differs significantly from the population mean.

We are provided that the sample is randomly selected and we assume that it has come from a normally distributed population.

### Step 3: Decide the significance level

The level of significance (alpha) as 0.05.

### Step 4: Calculate the p-value

````python
hyp_mean = 150
t_stat, p_value = stats.ttest_1samp(df['Unpolished '], hyp_mean, alternative='less') # alternative argument is used to specify the tail of the test.

print("Test Statistic =",t_stat)
print("p-value =",p_value)
````

### Step 5: Compare the p-value with $\alpha$

````python
# print the conclusion based on p-value
if p_value < 0.05:
    print(f'As the p-value {p_value} is less than the level of significance, we reject the null hypothesis.')
else:
    print(f'As the p-value {p_value} is greater than the level of significance, we fail to reject the null hypothesis.')
````

### Step 6: Draw Inference

Since the p-value is less than the 5% significance level, we reject the null hypothesis. Hence, Zingaro's thinking is justified.

**Paired t-test**

**Is the mean hardness of the polished and unpolished stones the same?**

###Step 1: Define null and alternative hypothesis

$H_0:$  The mean Brinell's hardness of polished and unpolished stones are equal.

$H_a:$ The mean Brinell's hardness of polished and unpolished stones are not equal.

### Step 2: Select Appropriate test

This is a problem, concerning two population means. A **paired t-test** is an appropriate test here.

### Step 3: Decide the significance level

The level of significance (alpha) as 0.05.

### Step 4: Calculate the p-value

````python
t_stat, p_value = stats.ttest_rel(df['Unpolished '],df['Treated and Polished']) # alternative argument is used to specify the tail of the test.

print("Test Statistic =",t_stat)
print("p-value =",p_value)
````

### Step 5: Compare the p-value with $\alpha$

````python
# print the conclusion based on p-value
if p_value < 0.05:
    print(f'As the p-value {p_value} is less than the level of significance, we reject the null hypothesis.')
else:
    print(f'As the p-value {p_value} is greater than the level of significance, we fail to reject the null hypothesis.')
````

### Step 6: Draw Inference

Since the p-value is less than the 5% significance level, we reject the null hypothesis. Hence, The mean Brinell's hardness of polished and unpolished stones are not equal.

**4.1 How does the hardness of implants vary depending on dentists?**

**Step 1: Define null and alternative hypotheses for One-Way ANONA**

$H_0$ : The mean hardness is equal across dentists.

$H_a$ : At least one dentist's mean hardness differ.

**Step 2: Select Appropriate test**

One-way ANOVA is an appropriate test here provided normality and equality of variance assumptions are verified.

**One-way ANOVA test**

In a one-way ANOVA test, we compare the means from several populations to test if there is any significance difference between them. The results from an ANOVA test are most reliable when the assumptions of normality and equality of variances are satisfied.

* For testing of normality, Shapiro-Wilk’s test is applied to the response variable.

* For equality of variance, Levene test is applied to the response variable.

**Shapiro-Wilk’s test**

We will test the null hypothesis

>$H_0:$ The data follow a normal distribution

against the alternative hypothesis

>$H_a:$ The data do not not follow a normal distribution

````python
Dentist_A1 = Alloy_1[Alloy_1['Dentist'] == 1]['Response']
Dentist_A2 = Alloy_1[Alloy_1['Dentist'] == 2]['Response']
Dentist_A3 = Alloy_1[Alloy_1['Dentist'] == 3]['Response']
Dentist_A4 = Alloy_1[Alloy_1['Dentist'] == 4]['Response']
Dentist_A5 = Alloy_1[Alloy_1['Dentist'] == 5]['Response']
````

````python
print(stats.shapiro([Dentist_A1]))
print(stats.shapiro([Dentist_A2]))
print(stats.shapiro([Dentist_A3]))
print(stats.shapiro([Dentist_A4]))
print(stats.shapiro([Dentist_A5]))
````

````python
Dentist_B1 = Alloy_2[Alloy_2['Dentist'] == 1]['Response']
Dentist_B2 = Alloy_2[Alloy_2['Dentist'] == 2]['Response']
Dentist_B3 = Alloy_2[Alloy_2['Dentist'] == 3]['Response']
Dentist_B4 = Alloy_2[Alloy_2['Dentist'] == 4]['Response']
Dentist_B5 = Alloy_2[Alloy_2['Dentist'] == 5]['Response']
````

````python
print(stats.shapiro([Dentist_B1]))
print(stats.shapiro([Dentist_B2]))
print(stats.shapiro([Dentist_B3]))
print(stats.shapiro([Dentist_B4]))
print(stats.shapiro([Dentist_B5]))
````

Though the test is failing here we are going with the assumption that the data follows normal distribution.


**Levene’s test**

We will test the null hypothesis

>$H_0:$ The variances are equal.

against the alternative hypothesis

>$H_a:$ The variances are not equal.

````python
print(stats.levene(Dentist_A1,Dentist_A2,Dentist_A3,Dentist_A4,Dentist_A5))
````

````python
print(stats.levene(Dentist_B1,Dentist_B2,Dentist_B3,Dentist_B4,Dentist_B5))
````

Since, p-value greater than 0.05 the variances are equal.

**Step 3: Decide the significance level**

As given in the problem statement, we select $\alpha = 0.05$

**Step 4: Calculate the p-value**

````python
formula = 'Response ~ C(Dentist)'
model = ols(formula, Alloy_1).fit()
aov_table = anova_lm(model)
print(aov_table)
````

````python
formula = 'Response ~ C(Dentist)'
model = ols(formula, Alloy_2).fit()
aov_table = anova_lm(model)
print(aov_table)
````

**Step 5: Draw inference**

Since p-value is greater than 0.05, we failed to reject null hypothesis. The mean hardnesses are equal across dentists.
